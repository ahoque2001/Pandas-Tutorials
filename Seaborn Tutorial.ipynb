# Correlation with Heatmap
## A correlation heatmap uses color coded cells, typically in a monochromatic scale to show a 2D correlation matrix(table) between two discrete dimension or event types. It is very important in feature selection

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

# Join plot
## A joint plot allows to study the relationship between 2 numeric variables. The central chart display their correlation. it is usually a scallar plot, a haxbin,a 2D histogram or a 2D density plot. It is help us to do Univariate analysis always two values

# Univariate analysis

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

# pair Plot: a pair plot is also known as scatter plot in which one variable in the same data row is matched with another variable's value,like this : pair plot is just elaboration on this . showing all variables paired with all the other variables.

```python
sns.pairplot(df)
```

## If you need classification with another category feature of independent variable you need to use "hue".

```python
df['tip'].value_counts()
```

```python
sns.pairplot(df,hue='smoker')
```

# Dist Plot
## Dist plot helps us to check the distribution of the column featuers

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
# Kernal density estimation
```

# Categorica plot
## Seaborn also help us to doing the analysis on Categorical Data points. In this section we will disclose about
## .Box Plot
## .violinplot
## .countplot
## .bar plot

```python
df['sex'].value_counts()
```

```python
sns.countplot('smoker',data=df)
```

```python
sns.countplot(x='day',data=df)
```

```python
# Bar plot

sns.barplot(y='total_bill',x='smoker',data=df)
```

```python
sns.barplot(y='total_bill',x='sex',data=df)
```

```python
df.head()
```

# Box plot
## A box plot and whisker plot (some time called a box plot) is a graph that present information from a five-number summey

```python
sns.boxplot('sex','total_bill',data=df)

# perentile 0%,25%,50% , 75% and 100% as shown below
```

```python
sns.boxplot(x='day',y='total_bill',data=df)
```

```python
sns.boxplot(x='day',y='total_bill',data=df,palette='rainbow')
```

```python
sns.boxplot(data=df,orient='v')
```

```python
# categorize my data based on some other categories

sns.boxplot(x='total_bill',y='day',hue='sex',data=df)
```

```python
sns.boxplot(x='total_bill',y='day',hue='smoker',data=df)
```

# violin Plot
## it is help us to see both the distribution of data in terms of Kernal density estimation and the boxplot

```python
sns.violinplot(x='total_bill',y='day',data=df,palette='rainbow')
```
